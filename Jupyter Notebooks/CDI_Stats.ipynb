```python
#Diane Stratton
#Hood College - CS 522 Fall 2018
#Project
#Notebook 2 - CDI statistics

#import required libraries
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
plt.rcParams['figure.figsize'] = (24,12)
plt.style.use('ggplot')
```

```python
#import modified CDI data file, and verify contents
cdi = pd.read_csv('CDI_2016_modif_v3.csv');
print(cdi.shape)
cdi.head()
```

```python
#Sort dataset by alc_binge attribute
cdi_sort = cdi.sort_values(by=['alc_binge'])
cdi_sort.head()
```

```python
#Display bar chart of sorted dataset
x_label = cdi_sort['state'].values
y_label = cdi_sort['alc_binge'].values

index = np.arange(len(x_label))
plt.bar(index, y_label)
plt.xlabel('State', fontsize=10)
plt.ylabel('Binge Drinking Incidence', fontsize=10)
plt.xticks(index, x_label, fontsize=5, rotation=30)
plt.title('Binge Drinking Incidence by State')
plt.show()
```
